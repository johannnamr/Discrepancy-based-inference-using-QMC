```
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import particles.qmc as qp # QMC points
from matplotlib.lines import Line2D
```

## Sampling uniforms

```
np.random.seed(11)
```

```
n = 2**9            # number of samples
```

MC points

```
u_mc = np.random.rand(n,2)
```

QMC points

```
u_qmc = qp.sobol(n,2,0)
```

## Sampling Gaussians

MC points

```
x_mc = stats.norm.ppf(u_mc, loc=0, scale=1)
```

QMC points

```
x_qmc = stats.norm.ppf(u_qmc, loc=0, scale=1)
```

## Plot

```
SMALL_SIZE = 20
MEDIUM_SIZE = 32
BIGGER_SIZE = 32

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)   # fontsize of the figure title
```

```
points = [u_mc, u_qmc, x_mc, x_qmc]
#fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(30,8), constrained_layout=True)
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,5.5), constrained_layout=True)
cmap = plt.get_cmap("tab10")

ax1.plot(points[0][:,0], points[0][:,1], 'o',color=cmap(0),label='MC')
#ax1.set_title("Uniform "+r'$\mathbb{U}$', pad=10, fontweight='bold')
ax1.set_title("Uniform "+r'$\mathbb{U}$')
ax1.set_xlabel(r'$u_1$')
ax1.set_ylabel(r'$u_2$')
#ax1.text(0.5, -0.35, 'Monte Carlo', ha="center", fontweight='bold')
ax1.set_xticks([0,0.5,1])
ax1.set_xlim(-0.05,1.05)
ax1.set_yticks([0,0.5,1])
ax1.set_ylim(-0.05,1.05)

ax2.plot(points[1][:,0], points[1][:,1], 'o',color=cmap(2),label='QMC')
#ax2.set_title("Uniform "+r'$\mathbb{U}$', pad=10, fontweight='bold')
ax2.set_title("Uniform "+r'$\mathbb{U}$')
ax2.set_xlabel(r'$u_1$')
#ax2.set_ylabel(r'$u_2$')
#ax2.text(0.5, -0.35, 'Quasi-Monte Carlo', ha="center", fontweight='bold')
ax2.set_xticks([0,0.5,1])
ax2.set_xlim(-0.05,1.05)
ax2.set_ylim(-0.05,1.05)
ax2.set_yticklabels([])

ax3.plot(points[2][:,0], points[2][:,1], 'o',color=cmap(0))
#ax3.set_title("Gaussian "+r'$\mathbb{P}_\theta$', pad=10, fontweight='bold')
ax3.set_title("Gaussian "+r'$\mathbb{P}_\theta$')
ax3.set_xlabel(r'$x_1$')
ax3.set_ylabel(r'$x_2$')
#ax3.text(0, -4.8, 'Monte Carlo', ha="center", fontweight='bold')
ax3.set_xticks([-2,0,2])
ax3.set_xlim(-3.1,3.1)
ax3.set_yticks([-2,0,2])
ax3.set_ylim(-3.1,3.1)

ax4.plot(points[3][:,0], points[3][:,1], 'o',color=cmap(2))
#ax4.set_title("Gaussian "+r'$\mathbb{P}_\theta$', pad=10, fontweight='bold')
ax4.set_title("Gaussian "+r'$\mathbb{P}_\theta$')
ax4.set_xlabel(r'$x_1$')
#ax4.set_ylabel(r'$x_2$')
#ax4.text(0, -4.85, 'Quasi-Monte Carlo', ha="center", fontweight='bold')
ax4.set_xticks([-2,0,2])
ax4.set_xlim(-3.1,3.1)
ax4.set_yticks([-2,0,2])
ax4.set_yticklabels([])
ax4.set_ylim(-3.1,3.1)

#handles = [Line2D([0], [0], color=cmap(0), marker='o', markersize=10,linestyle='None',label='MC'),Line2D([0], [0], color=cmap(2), marker='o', markersize=10,linestyle='None',label='QMC')]
#plt.legend(bbox_to_anchor=(0,1.02,1,0.2), loc="lower left",mode="expand", borderaxespad=0, ncol=3)
#plt.legend(handles=handles,loc=7,bbox_to_anchor=(0.5,0.5,0.5,0.2),mode="expand", borderaxespad=0)

plt.savefig('plotQMC.png',bbox_inches = 'tight', pad_inches = .1)
plt.show() 
```
